```python
!apt-get update -y
!apt-get install -y tesseract-ocr
!pip -q install pytesseract pandas openpyxl Pillow

import re
import pandas as pd
import pytesseract
from PIL import Image
from google.colab import files


def normalize_to_10_digits(s: str) -> str | None:
    """Keep only digits and return the last 10 (Indian mobile style)."""
    digits = re.sub(r"\D", "", str(s))
    if len(digits) < 10:
        return None
    return digits[-10:]


PHONE_PATTERN = re.compile(r"(?:\+91\s*)?\b\d{5}\s*\d{5}\b")


uploaded_excel = files.upload()
excel_name = next(iter(uploaded_excel.keys()))

uploaded_imgs = files.upload()

df = pd.read_excel(excel_name, engine="openpyxl")
excel_nums = set()

for val in df.iloc[:, 0].tolist():
    n = normalize_to_10_digits(val)
    if n:
        excel_nums.add(n)

img_nums = set()

for fname in uploaded_imgs.keys():
    img = Image.open(fname)

    text = pytesseract.image_to_string(img)

    for match in PHONE_PATTERN.findall(text):
        n = normalize_to_10_digits(match)
        if n:
            img_nums.add(n)

missing = sorted(img_nums - excel_nums)

print(f"Numbers found in screenshots but NOT in Excel: {len(missing)}\n")
for n in missing:
    print(n)
```
